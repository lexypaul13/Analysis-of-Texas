# Analysis of Texas

The purpose of this analysis is to conduct a profile of Texas in terms of Rent Price, Crime and Job Market Size

# Apartment Data

The data is collected from quandl and plotted with matplotlib

```python
import quandl
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
api_key = open('quandlapikey.txt','r').read() 
```

The code collects the rental data for apartments in texas starting from the year 1990

```python
api_key = open('quandlapikey.txt','r').read()
df_rent = quandl.get("NAHB/MFVI_RVCP", authtoken=api_key,start_date="1990-01-01")

df_rent = df_rent.drop(["Multifamily Vacancy Index", 'Class A Apartments', "Class B Apartments"], axis=1)

df_rent.rename(columns={'Class C Apartments':'Rent For Apartment In Texas'}, inplace=True)
```

The data is now plotted using matplotlib

```python
import matplotlib.pyplot as plt
from matplotlib import style


df_rent.plot()
plt.legend()
plt.title("Rent For Apartment In Texas")
plt.ylabel("price of rent")
plt.show()
```

# Job Market Size 

The code below collects the labour market data in texas starting from the year 1990

```python
api_key = open('quandlapikey.txt','r').read()
df_job = quandl.get("FRED/TX41425000M175FRBDAL", authtoken=api_key,start_date="1990-01-01")
df_job.rename(columns={'Value':'Job Market'}, inplace=True)
```

The data is now plotted 

```python
import matplotlib.pyplot as plt
from matplotlib import style
df_job.plot()
plt.title("Job Market In Texas")
plt.ylabel("size of workforce")
plt.legend()
plt.show()
```
